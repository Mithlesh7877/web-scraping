```python
import requests
from bs4 import BeautifulSoup
```

```python
base='https://editorial.rottentomatoes.com/guide/140-essential-action-movies-to-watch-now/2/'
```

```python
res=requests.get(base)
```

```python
res.status_code
```

```python
html=res.content
```

```python
soup=BeautifulSoup(html,'lxml')
```

```python
with open(r'C:\Users\My PC\Desktop\Machine Learning\rotten.html','wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
divs=soup.find_all('div',{'class':'col-sm-18 col-full-xs countdown-item-content'})
```

```python
head=[l.find('h2')for l in divs]
```

```python
title=[t.find('a').string for t in head]
```

```python
year=[s.find('span',class_='start-year').string for s in head]
```

```python
# Cast
```

```python
ca=[div.find('div',{'class':'cast'}) for div in divs]
ca
```

```python
ca[0]

```

```python
cast_links=ca[0].find_all('a')
cast_links
```

```python
cast_names=[link.string for link in cast_links]
```

```python
cast_names
```

```python
cast=", ".join(cast_names)
```

```python
cast
```

```python
#for loop
cast_f=[]
for i in ca:
    ca_links=i.find_all('a')
    cast_name=[link.string for link in ca_links]
    c=", ".join(cast_name)
    cast_f.append(c)
```

```python
cast_f
```
